```python
from map_boxes import mean_average_precision_for_boxes
import pandas as pd

solution_DB = pd.read_csv('./DB1_defect_points.csv')
solution_DB.columns = solution_DB.columns.str.replace(' ','')
solution_DB
```

```python
IDA =[] 
nameA = []
xminA = []
yminA = []
xmaxA = []
ymaxA = []

for ind in range(0,len(solution_DB)):
    
    IDA.append(solution_DB.file_name[ind])
    nameA.append('defect')

    xmin = int(min(solution_DB.x_start[ind],solution_DB.x_end[ind]))
    ymin = int(min(solution_DB.y_start[ind],solution_DB.y_end[ind]))
    xmax = int(max(solution_DB.x_start[ind],solution_DB.x_end[ind]))
    ymax = int(max(solution_DB.y_start[ind],solution_DB.y_end[ind]))

    xminA.append(xmin/1024)
    yminA.append(ymin/1024)
    xmaxA.append(xmax/1024)
    ymaxA.append(ymax/1024)

annotations = pd.DataFrame({
    "ImageID": IDA,
    "LabelName": nameA,
    "XMin": xminA,
    "XMax": xmaxA,
    "YMin": yminA,
    "YMax": ymaxA
})
annotations['ImageID'] = annotations['ImageID'].replace(' ', '', regex=True)
# annotations.to_csv('./annotations.csv', index = False)
annotations = annotations.sort_values(by='ImageID', ascending=True)
annotations
```

```python
results_DB = pd.read_csv('./Coordonnes_anomaly.csv')
results_DB['Conf'] = 1.0
results_DB.columns = results_DB.columns.str.replace(' ','')
results_DB['defect_numb'] = results_DB['defect_numb'] + 1
results_DB['LabelName'] = 'defect'
results_DB
```

```python
IDA =[] 
nameA = []
confA = []
xminA = []
yminA = []
xmaxA = []
ymaxA = []

for ind in range(0,len(results_DB)):
    
    IDA.append(results_DB.file_name[ind])
    nameA.append('defect')
    confA.append(results_DB.Conf[ind])

    xmin = int(min(results_DB.x_start[ind],results_DB.x_end[ind]))
    ymin = int(min(results_DB.y_start[ind],results_DB.y_end[ind]))
    xmax = int(max(results_DB.x_start[ind],results_DB.x_end[ind]))
    ymax = int(max(results_DB.y_start[ind],results_DB.y_end[ind]))

    xminA.append(xmin/1024)
    yminA.append(ymin/1024)
    xmaxA.append(xmax/1024)
    ymaxA.append(ymax/1024)

detections = pd.DataFrame({
    "ImageID": IDA,
    "LabelName": nameA,
    "Conf": confA,
    "XMin": xminA,
    "XMax": xmaxA,
    "YMin": yminA,
    "YMax": ymaxA
})
detections['ImageID'] = detections['ImageID'].replace(' ', '', regex=True)
# detections.to_csv('./detections.csv', index = False)
detections = detections.sort_values(by='ImageID', ascending=False)
detections
```
